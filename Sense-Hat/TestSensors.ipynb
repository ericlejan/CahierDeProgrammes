# Test des capteurs du Sense-Hat

```python
from sense_hat import SenseHat

sense = SenseHat()

sense.set_rotation(180)
sense.low_light = True

humidity = sense.get_humidity()
print("Humidity: %s %%rH" % humidity)
sense.show_message("Humidity: %.2f %%rH" % humidity)

temp = sense.get_temperature()
print("Temp. : %s C" % temp)
sense.show_message("Temp. : %.2f C" % temp)

tempS = sense.get_temperature_from_humidity()
print("TempS. : %s C" % tempS)
sense.show_message("TempS. : %.2f C" % tempS)

tempP = sense.get_temperature_from_pressure()
print("TempP. : %s C" % tempP)
sense.show_message("TempP. : %.2f C" % tempP)

pressure = sense.get_pressure()
print("Press. : %s hPa" % pressure)
sense.show_message("Press. : %d hPa" % pressure)

```

```python
from sense_hat import SenseHat
import time, math

sense = SenseHat()
sense.set_imu_config(False, False, True)  # accelerometer only
angle = 0

while True :
    rawg = sense.get_accelerometer_raw()
    angle = math.pi * math.atan2(rawg.get('y'),rawg.get('x')) / 180
    print("x: {x}, y: {y}, z: {z}".format(**rawg), "Angle : %.2f" %angle)
```
