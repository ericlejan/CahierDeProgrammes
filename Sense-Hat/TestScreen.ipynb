```python
# -*- coding: utf-8 -*-

from sense_hat import SenseHat
import time, math

sense = SenseHat()
sense.set_rotation(180)
sense.set_imu_config(False, False, True)  # accelerometer only
sense.low_light = True

# Creating colors
dred = (128,0,0)
red = (255, 0, 0)
yellow = (255, 255, 0)
green = (0, 255, 0)
dgreen = (0, 128, 0)
cyan = (0, 255, 255)
blue = (0, 0, 255)
dblue = (0, 0, 255)
off = (0,0,0)

jauge = [
    dblue, blue, cyan, dgreen, green, yellow, dred, red
]
negjauge = [
    red, dred, yellow, green, dgreen, cyan, blue, dblue
]


#lighting the matrix
#for i in range(0,8,1) :
#    sense.set_pixel(0, i, red)
#    sense.set_pixel(1, i, dred)
#    sense.set_pixel(2, i, yellow)
#    sense.set_pixel(3, i, green)
#    sense.set_pixel(4, i, dgreen)
#    sense.set_pixel(5, i, cyan)
#    sense.set_pixel(6, i, blue)
#    sense.set_pixel(7, i, dblue)
#time.sleep(1)
#sense.clear()  # no arguments defaults to off

#lighting the matrix
for i in range(0,8,1) :
    for j in range(0,8,1) :
        sense.set_pixel(j, i, jauge[j])
        time.sleep(0.02)
    time.sleep(0.1)
#sense.clear()  # no arguments defaults to off

angoff = 33.5 # angular offset of accel vector
scale = 6

#for i in range(0,100,1) :
while True :
    matrix = sense.get_pixels()
    rawg = sense.get_accelerometer_raw()
    time.sleep(0.005)
    angle = math.atan2(rawg.get('y'),rawg.get('x')) / math.pi * 180 - angoff
    num = (math.floor(scale * angle))
    if (num > 8) :
        num=0
    if (num < -8) :
        num=0
#    print("angle : %.3f" %angle, "num : %.3f" %num)
    if (num > 0) :
        for l in range(0,7,1) :
            for k in range(0,8,1) :
                sense.set_pixel(k,l,matrix[k + 8*(l+1)])
        for k in range(0,8,1) :
            sense.set_pixel(k,7,off)
        for j in range(0,num,1) :
#            sense.set_pixel(j, 7, jauge[j])
            sense.set_pixel(j, 7, red)
    else :
        num = -num
        for l in range(0,7,1) :
            for k in range(0,8,1) :
                sense.set_pixel(k,l,matrix[k + 8*(l+1)])
        for k in range(0,8,1) :
            sense.set_pixel(k,7,off)
        for j in range(0,num,1) :
#            sense.set_pixel(j, 7, negjauge[j])
            sense.set_pixel(j, 7, blue)
#    time.sleep(1)

sense.clear()  # no arguments defaults to off
   
```
